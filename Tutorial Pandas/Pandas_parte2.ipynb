# Função Maps
”Map” é um termo, emprestado da matemática, para uma função que pega um conjunto de valores e os “mapeia” para um novo conjunto de valores. Na ciência de dados, geralmente precisamos alterar a representação dos dados, o método maps pode ser uma ferramenta extremamente útil para isso.
    
Existem dois tipos de “mapeamento” que usaremos com frequência. map() é um pouco mais simples, vamos supor que desejássemos reorganizar a o preço dos vinhos. Podemos fazer isso da seguinte maneira:


```python
wine_price_mean = wine.price.mean()
wine.price.map(lambda p: (p - wine_price_mean))
```

```python
#Note que não houve modificação no DataFrame
wine.head()
```

A função que você passa para o método map() é aplicada a todos os valores da Série e retorna uma nova série com os dados transformados. 

O método apply() é equivalente se quisermos transformar um DataFrame inteiro chamando um método personalizado a cada linha.


```python
def remean_price(df):
    df.price = df.price - wine_price_mean
    return df
wine.apply(remean_price, axis='columns')
```

Se nós tivéssemos chamado wine.apply() com axis=’index’, então teríamos passado uma função para transformar cada linha, nós precisamos dar uma função para transformar cada coluna.

Note que tanto map() e apply() retornam novas Series e DataFrames, não modificando os originais.

A biblioteca pandas permite muitas opções de “mapeamento” já embutidas, por exemplo, aqui temos uma maneira mais simples para remanejar os itens das colunas:


```python
wine_price_mean = wine.price.mean()
wine.price - wine_price_mean
```

Nesse código nós estamos performando uma operação entre muitos valores, do lado esquerdo temos o número da operação e do lado direito temos o “valor médio”. A biblioteca pandas olha para a expressão e entende que devemos subtrair a média dos valores, para cada valor no conjunto de dados.

Pandas também entenderá o que ele deve fazer entre operações quando as Séries apresentam o mesmo comprimento. Por exemplo, uma maneira fácil de combinar o label e a provincia seria:


```python
wine.label.apply(str) + " - " + wine.province.apply(str)
```

Esses operadores são mais rápido do que os métodos map() ou apply() porque eles já têm acelerações embutidas no pandas. Todos os operadores padrões (nativos do python) funcionam dessa maneira.

Entretanto, esses operadores não são tão flexíveis quanto map() e apply(), os quais podem ser usados para fazer coisas mais avançadas, como aplicar condicionais, que não podem ser realizadas adições e subtrações sozinhas.


# Agrupando e classificando
<h3>Introdução</h3>
O “mapeamento” permite que transformemos os dados de um DataFrame ou uma Série um a um para uma coluna inteira. Entretanto, em muitos casos precisamos agrupar os dados, e então fazer alguma coisa com o grupo que os dados estão.

Assim como você aprenderá, nós faremos isso utilizando a operação groupby(). Nós também iremos cobrir alguns tópicos adicionais, como formas mais complexas de indexar seus DataFrames, além de como classificar seus dados.

<h3>Análise de grupo</h3>
	Uma função que nós temos usado de maneira frequente é a value_counts(). Nós podemos replicar o que value_counts() faz usando o seguinte:


```python
wine.groupby('price').price.count()
```

groupby() criou um grupo das wine que atribuiu os mesmos valores em pontos aos vinhos indicados. 

Então, para cada um desses grupos, nós contamos quantas vezes ela apareceu. value_counts() é apenas um atalho para esta operação groupby().

Nós podemos usar qualquer uma das funções de resumo que nós já usamos com esses dados. Por exemplo, para pegar o vinho mais barato em cada país, nós podemos fazer o seguinte:


```python
wine.groupby('country').price.min()
```

Você pode pensar em cada grupo que geramos como uma fatia do nosso DataFrame contendo apenas dados com valores correspondentes. Esse DataFrame pode ser acessado diretamente por nós utilizando o método apply(), e podemos manipular os dados da maneira que acharmos melhor.

Por exemplo, aqui temos um exemplo de uma maneira de obter a quantidade de classificações feitas por cada revisor


```python
pd.DataFrame(wine.groupby('taster_name').apply(lambda df: df.quality.value_counts()))
```

Para cada controle ainda mais refinado você também pode agrupar por mais de uma coluna, por exemplo, eis como escolheríamos o melhor vinho (maior qualidade) por país e província.


```python
wine.groupby(['country', 'province']).apply(lambda df: df.loc[df.quality.idxmax()])
```

# Tipos de dados e valores faltantes
<h3>Tipos de dados</h3>

Às vezes o tipo de determinado dado não é adequado, para isso devemos modificar seu tipo. No Pandas tempos a função dtype() que fornece o tipo de dado.

Utilizando o exemplo dos vinhos, se quisermos obter o tipo de dados de todas as colunas podemos utilizar:


```python
wine.dtypes
```

Caso queiramos obter apenas o tipo de dado de dada coluna:


```python
wine.price.dtype
```

Note que valores do tipo inteiro são classificados como int64, valores do tipo ponto flutuante são classificados como float64, e strings são classificadas como objects!

Caso faça sentido para o Pandas, pode-se alterar o tipo de dado utilizando o método astype. Por exemplo:


```python
wine.price.astype('int64')
```

# Renomeando
<h3>Introdução</h3>

Algumas vezes os dados vêm com nomes de colunas ou índices que podem ser confusas ou difíceis de trabalhar. Para isso o pandas conta com alguns métodos que permitem a renomeação de colunas e índices.
Renomeando

O método rename()  permite a modificação dos nomes das colunas e dos índices (rótulos). Por exemplo, para modificar a coluna price por score:


```python
wine.rename(columns={'price':'preço'})
```

Para modificar os índices:

```python
wine.rename(index={0:'index0',1:'index1'})
```

Renomear colunas é algo bem comum, já renomear índices não tanto, por isso em grande parte das vezes é mais conveniente usar o método  set_index().

Tanto os índices das linhas como os das colunas podem ter seu próprio atributo nome. O método rename_axis() pode ser utilizado para modificar esses nomes.
