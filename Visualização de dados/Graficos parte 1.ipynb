```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header=None)
iris.columns = columns=['sepal_length', 'sepal_width','petal_length', 'petal_width','class']
```

```python
iris
```

```python
plt.title('Boxplot Sepal Length')
plt.boxplot(iris['sepal_length']);
```

```python
plt.bar(iris['class'],height=50);
```

```python
iris['class'].value_counts()
```

```python
#gráfico de dispersão
sns.scatterplot(data=iris ,x='sepal_length', y='sepal_width');
```

```python
#gráfico de regressão
sns.regplot(data=iris ,x='sepal_length', y='petal_length');
```

```python
sns.violinplot(data=iris, y='sepal_length',x='class');
```

```python
sns.boxplot(data=iris, y='sepal_length',x='class');
```

```python
plt.figure(figsize=[20,10])
sns.boxplot(x="petal_length", y="sepal_length",
            hue="class", palette=["b", "r","g"],
            data=iris);
```

```python
#Countplot
plt.figure(figsize=[20,7])
plt.title('Countplot sepal length')
sns.countplot(data=iris,x='sepal_length');
```

```python
iris.sepal_length.min(),iris.sepal_length.max(),(iris.sepal_length.max()-iris.sepal_length.min())/12
```

```python
iris2 = iris.copy()
iris2['sepal_length'] = pd.cut(iris2['sepal_length'],12,labels=[round(i,2) for i in np.arange(4.3,7.9,0.31)])
```

```python
plt.figure(figsize=[9,7])
sns.countplot(data=iris2,x='sepal_length');
```

```python
sns.jointplot(data=iris, x='sepal_length', y='sepal_width', hue='class');
```

```python
g = sns.JointGrid(data=iris, x='sepal_length', y='sepal_width', hue='class')
g.plot(sns.scatterplot, sns.histplot);
```

```python
k = sns.JointGrid(data=iris, x='sepal_length', y='sepal_width', hue='class')
k.plot(sns.scatterplot, sns.boxplot);
```

```python
df=pd.DataFrame({'x': range(1,51), 'y': np.random.randn(50), 'k': np.random.randn(50), 'l':np.random.randn(50)})
plt.plot(df.x,df.y);
```

```python
columns= ['y','k','l']
```

```python
plt.figure(figsize=[18,7])
for col in columns:
    plt.plot(df['x'], df[col], marker='o', linewidth=1.8, alpha=1, label=col)
plt.xlabel('Time')
plt.ylabel('Random numbers')
plt.legend();
```
